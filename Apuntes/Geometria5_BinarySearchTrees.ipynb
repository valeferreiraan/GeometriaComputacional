Ordenar x usando un árbol binario.

- **Preorder:** `(Raíz, hijo izq, hijo der)` -> típicamente para copiar árboles
- **Inorder:** `(hijo izq, raíz, hijo der)` -> va a servir en nuestro caso para leer los elementos de un arbol y que ningun caso sea decreciente cuando se lea.
- **PosOrder:** `(hijo izq, hijo der, raíz)` -> proceso típico para borrar elementos según árbol.

- **Profundidad del árbol:** Como se tieene planeado, la profundidad es de 4.

Podemos calcular si está balanceado o no. Lo que se espera para un árbol balanceado sea 1 o 0.


````python
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

def insertBST(root, node): #insertar binary search tree
  if(node.data < root.data):
    if root.left is None:
      root.left = node
    else:
      insertBST(root.left, node)
  else:
    if root.right is None:
      root.right = node
    else:
      insertBST(root.right, node)
````

````python
x =[25,75,50,10,100,15,0,30,200,40]
````

````python
xc = x.copy()
root = Node(xc[0])
xc.pop(0)

while len(xc):
  insertBST(root, Node(xc[0]))
  xc.pop(0)

print(root.data)
print(root.right.data)
print(root.left.data)
````

````python
def preorder(node):
  if not node:
    return
  print(node.data)
  preorder(node.left)
  preorder(node.right)
````

````python
preorder(root)
````

````python
def inOrder(root):
  inorderArr = []
  if root.left is not None:
    leftInorderArr = inOrder(root.left)
    inorderArr.extend(leftInorderArr)

  inorderArr.append(root.data)

  if root.right is not None:
    rightInorderArr = inOrder(root.right)
    inorderArr.extend(rightInorderArr)

  return inorderArr
````

````python
inOrder(root)
````

````python
def depth(root):
  if root is None:
    return -1
  leftAns = depth(root.left)
  rightAns = depth(root.right)

  return max(leftAns, rightAns) + 1
````

````python
depth(root)
````

````python
depth(root.left)
````

````python
depth(root.right)
````

Ahora, vamos a construir el árbol balanceado a partir del arreglo ordenado.

````python
def balancedTree(sortedArr):
  if not sortedArr:
    return None

  mid = int(len(sortedArr)/2)
  newRoot =  Node(sortedArr[mid])
  newRoot.left = balancedTree(sortedArr[:mid])
  newRoot.right = balancedTree(sortedArr[mid+1:])

  return newRoot
````

````python
root2 = balancedTree(inOrder(root))
root2.data
````

````python
depth(root2)
````

````python
depth(root2.left)
````

````python
depth(root2.right)
````

### **Tarea: Establecer una relacion de orden para puntos en $R^2$**
modificar el arbol binario para que ordene los puntos ordenados segun una relacion de orden.

La relación de orden que se seguirá es la  siguiente:
> * Dados dos puntos $p,q \in R^2$:
>   * $p<q \iff p_y > q_y \lor (p_y = q_y \land p_x < q_x)$

````python
import numpy as np
import matplotlib.pyplot as plt

N = 10 #cantidad de numeros
np.random.seed(22)
x = np.random.randint(low=0, high=25, size = N)
y = np.random.randint(low=0, high=25, size = N)


points = []
for i in range(N):
  points.append([x[i], y[i]])
print(points)

````

````python
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None
````

````python
def sortpoints(root, node): #insertar binary search tree
  if(root.data[1] > node.data[1]):
    if root.right is None:
      root.right = node
    else:
      sortpoints(root.right, node)
  elif(node.data[1] == root.data[1] and node.data[0] > root.data[0]):
    if root.right is None:
      root.right = node
    else:
      sortpoints(root.right, node)
  else:
    if root.left is None:
      root.left = node
    else:
      sortpoints(root.left, node)
````

````python
pc = points.copy()
root = Node(pc[0])
pc.pop(0)
print(root.data)
````

````python
while len(pc):
  sortpoints(root, Node(pc[0]))
  pc.pop(0)

print(root.data)
print(root.right.data)
print(root.left.data)

````

````python
sorted = inOrder(root)
sorted
````
