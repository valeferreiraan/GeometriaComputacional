```python
def rotaciones(x1,y1, x2, y2):
  determinante = x1*y2-x2*y1
  if determinante > 0:
    print('P1 esta rotado en sentido horario con respecto a P2')
  elif determinante < 0:
    print('P1 esta rotado en sentido anti-horario con respecto a P2')
  else:
    print('Los vectores son colineales')
```

```python
#alternativa de solucion
import numpy as np
import matplotlib.pyplot as plt

class Point:
  #constructor
  def __init__(self, x, y, name):
    self.x = x
    self.y = y
    self.name = name

  #metodos de la clase
  def __repr__(self):
    return(f"({self.x}, {self.y})")

  def __str__(self):
    return(self.name)
```

```python
def cross(p0, p1, p2): #origen comun, p1 , p2
  return((p1.x - p0.x)*(p2.y- p0.y)- (p2.x - p0.x)*(p1.y-p0.y))
```

##**Pregunta 3**
¿Cómo responder la pregunta 3?

- que d0 y d1 tengan signos contrarios $(d_0*d_1< 0)$ -> **punto de partida**
- entonces uno debe girar en sentido horario y otro anti horario
- sacarle provecho al producto cruz

Para que haya intersección debe cumplir:
$$ d_0 * d_1 < 0 $$
&
$$ d_2 * d_3 < 0 $$
si esto se cumple, hay intersección en un **punto intermedio de los segmentos**

Si $d_3 = 0$ y $p_3 \in \bar{p_0 p_1} ⇒$ hay intersección en P3

Si los segmentos se cruzan, dónde se cruzan?

los segmentos que se estan considerando son intervalos cerrados, es decir, se tienen en cuenta los extremos

Para hacer el análisis de donde se cruzan, vamos a hacerlo con combinación convexa:
$$\bar{P_0P_1} = \alpha P_0 + (1- \alpha) P_1, ~ \alpha \in [0,1]$$

$$\bar{P_2 P_3} = \beta \bar{P_2} + (1-\beta) \bar{P_3},~~~ \beta \in [0,1]$$



**RESOLVER**
$$ \alpha \bar{P_0} + (1-\alpha) \bar{P_1} = \beta \bar{P_2} + (1-\beta)\bar{P_3} $$

```python

#Determinar si dos segmentos de línea se interceptan
def on_segment(pi, pj, pk):
  if min(pi.x, pj.x) <= pk.x <= max(pi.x, pj.x) and min(pi.y, pj.y) <= pk.y <= max(pi.y, pj.y):
    return True
  else:
    return False

def cutPoint (p0, p1, p2, p3):
  A = np.array([[p0.x - p1.x, p3.x-p2.x], [p0.y-p1.y, p3.y-p2.y]])
  b = np.array([p3.x-p1.x, p3.y-p1.y])
  sol = np.linalg.solve(A,b)
  pc = [sol[0]*p0.x + (1-sol[0])*p1.x, sol[0]*p0.y + (1-sol[0])*p1.y]
  return pc

def segmentsInt (p1, p2, p3, p4):
  d1 = cross(p3, p4, p1)
  d2 = cross(p3, p4, p2)
  d3 = cross(p1, p2, p3)
  d4 = cross(p1, p2, p4)
  if ((d1>0 and d2>0) or (d1<0 and d2>0)) and ((d3>0 and d4<0) or (d3<0 and d4>0)):
    pc = cutPoint(p1,p2,p3,p4)
    return True, pc
  elif (d1 == 0) and on_segment(p3, p4, p1):
    pc = cutPoint(p1,p2,p3,p4)
    return True, pc
  elif (d2 == 0) and on_segment(p3, p4, p2):
    pc = cutPoint(p1,p2,p3,p4)
    return True, pc
  elif (d3 == 0) and on_segment(p1, p2, p3):
    pc = cutPoint(p1,p2,p3,p4)
    return True, pc
  elif (d4 == 0) and on_segment(p1, p2, p4):
    pc = cutPoint(p1,p2,p3,p4)
    return True, pc
  else:
    return False
```

```python
pu0 = Point(1,1, 'P1')
pu1 = Point(3,4, 'P2')
pu2 = Point(2,4, 'P3')
pu3 = Point(3,1, 'P4')


plt.figure(figsize = (3,3))
# Asi se grafica :)
plt.arrow(pu0.x, pu0.y, pu1.x - pu0.x, pu1.y - pu0.y, head_width = 0.05, head_length = 0.05,
          color = "blue", label = f"{p1}-{p0}")
plt.arrow(pu2.x, pu2.y, pu3.x - pu2.x, pu3.y - pu2.y, head_width = 0.05, head_length = 0.05,
            color = "purple", label = f"{p2}-{p0}")

r = segmentsInt(pu0,pu1,pu2,pu3)
if r[0] is True:
  print("Hay una intersección")
  print(r[1])
else:
  print("No hay intersecciones")
plt.scatter(r[1][0], r[1][1])
plt.show()
```

```python
P1 = Point(-1,1, 'P1')
P2 = Point(1,-1, 'P2')
P3 = Point(1,1, 'P3')
P4 = Point(-1,-1, 'P4')

plt.figure(figsize = (3,3))
plt.arrow(P1.x, P1.y, P2.x, P2.y,  color = "blue")
plt.arrow(P3.x, P3.y, P4.x, P4.y,  color = "red")

r = segmentsInt(P1,P2,P3,P4)
if r[0] is True:
  print("Hay una intersección")
else:
  print("No hay intersecciones")
plt.scatter(r[1][0], r[1][1])
plt.show()
```
