```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cross(x0, y0, x1, y1, x2, y2):
  return ((x1-x0)*(y2-y0) - (x2-x0)*(y1-y0))
```

```python
def isConvex(pol):
  cont = 0
  der = 0
  izq = 0

  while cont <= len(pol)-3:
    d = cross(pol[cont][0], pol[cont][1], pol[cont+1][0], pol[cont+1][1], pol[cont+2][0], pol[cont+2][1])
    if d <= 0:
      der += 1
      print("der")
    elif d > 0:
      izq += 1
      print("izq")
    cont += 1
  if der > 0 and izq == 0:
    print("Es convexo")
  elif der == 0 and izq > 0:
    print("Es convexo")
  else:
    print("no es convexo")

```

```python
def poli(points):
  n = len(points)
  x = []
  y = []
  for i in range(n):
    x.append(points[i][0])
    y.append(points[i][1])
  plt.scatter(x,y,color = 'purple')
  pol = []
  for i in range(n):
    pol.append(points[i])
  pol.append(points[0])
  xc = []
  yc = []
  for i in range(len(pol)):
    xc.append(pol[i][0])
    yc.append(pol[i][1])
  plt.plot(xc, yc,color = 'purple')
  plt.show()
  isConvex(pol)

```

```python
P1 = [(2.2, 1.6), (2.0, 0.2), (1.5, 0.3), (0.7, 1.3), (0.5, 1.7)]
poli(P1)
```

```python
P2 = [(1.4, 1.2), (2.2, 1.6), (2.0, 0.2), (1.5, 0.3), (0.7, 1.3)]
poli(P2)
```
